# Train 1 Merge

```python
import os
import pandas as pd
import time

#read the path
file_path = "D:/Python Codes/PBL Project/CAX_Train_1 csv"
os.chdir(file_path)
#list all the files from the directory
file_list = os.listdir(file_path)
file_list
```

```python

df_append = pd.DataFrame()
#append all files together
for file in file_list:
            df_temp = pd.read_csv(file)
            df_append = df_append.append(df_temp, ignore_index=True)
df_append
df_append.ACID.unique()
```

```python

os.chdir("C:/Users/vraja/PBL Project")
df_append.to_csv("Train 1.csv")
```

# Final Merge (Master Dataset)

```python

df_master = pd.DataFrame()
import os
import glob
import pandas as pd
#list all csv files only
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
#append all files together
for file in csv_files:
            df_temp = pd.read_csv(file)
            df_master = df_master.append(df_temp, ignore_index=True)

print(df_master.ACID.unique())
df_master.to_csv("Master Data.csv",index=False)
```
