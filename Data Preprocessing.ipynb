# Combining Datetime Values

```python
df["Year"]=df["Year"].astype(str)
df["Month"]=df["Month"].astype(str)
df["Day"]=df["Day"].astype(str)
df["Hour"]=df["Hour"].astype(str)
df["Minute"]=df["Minute"].astype(str)
df["Second"]=df["Second"].astype(str)
```

```python
df.rename(columns = {'Year':'year', 'Month':'month',
                              'Day':'day','Hour':'hour','Minute':'minute','Second':'second'}, inplace = True)
```

```python
df['datetime']=pd.to_datetime(df[['year','month','day','hour','minute','second']])
```

```python
df.drop(['year','month','day','hour','minute','second'],axis=1,inplace=True)
```
